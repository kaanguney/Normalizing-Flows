```python
import warnings 
warnings.filterwarnings("ignore", category=DeprecationWarning)
import numpy as np
import pandas as pd
import scipy # could be irrelevant, possibly needed for scatter_matrix 
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
%matplotlib inline
plt.style.use("seaborn")

import tensorflow as tf
print(tf.__version__) # has to be tensorflow 1.x for compatibility

import os 
os.environ["TF_CPP_MIN_LOG_LEVEL"]="3"
```

```python
tfd = tf.contrib.distributions
tfb = tfd.bijectors
```

```python
session = tf.compat.v1.InteractiveSession()
# if there exists a session, run session.close() first
tf.compat.v1.set_random_seed(42)
```

```python
batch_size=32
DTYPE=tf.float32
NP_DTYPE=np.float32
num_bijectors=8
```

```python
# print(os.getcwd())
data_dir = "/Users/kaanguney.keklikci/Desktop/Erasmus+/Heidelberg/data/"
data = pd.read_excel(data_dir+"prostate.xls", engine="xlrd")
data.head()
```

```python
data.isnull().sum()
```

```python
# impute missing values
data = data.dropna(subset = ["ekg", "age"])
data["sz"] = data["sz"].fillna(value = np.mean(data["sz"]))
data["sg"] = data["sg"].fillna(value = np.mean(data["sg"]))
data["wt"] = data["wt"].fillna(value = np.mean(data["wt"]))

# check if all are imputed 
assert(data.isnull().all().sum() == 0)
```

```python
data.shape, data.describe()
```

```python
data.drop(["patno", "sdate"], axis=1, inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
print(data.rx.unique())
print(data.ekg.unique())
print(data.pf.unique())
data = pd.get_dummies(data)
```

```python
columns = data.columns 
data = StandardScaler().fit_transform(data) # returns values, not dataframe
data = pd.DataFrame(data,columns=columns) 
data.head()
```

```python
# visualize the figure as line plot for each row 
# remove legend, too long
data.T.plot(figsize=(10,5))
plt.xticks(rotation=45)
plt.legend().remove()
plt.show() 
```

```python
# alternative visualization: plot densities of hg, sz, wt
scatter_matrix(data[["hg","sz","wt"]],alpha=0.5,figsize=(5,5),diagonal="kde")
plt.show()
```

```python
# experiment on 2 dimensions only 
data = data[["wt","sz"]]
scatter_matrix(data,alpha=0.5,figsize=(5,5),diagonal="kde")
plt.show()
print(data.shape)
```

```python
dataset = tf.data.Dataset.from_tensor_slices(data.astype(NP_DTYPE))
dataset = dataset.repeat()
dataset = dataset.shuffle(buffer_size=data.shape[0])
dataset = dataset.prefetch(2*batch_size)
dataset = dataset.batch(batch_size)
data_iterator = tf.compat.v1.data.make_one_shot_iterator(dataset)
x_samples = data_iterator.get_next()
```

```python
base_dist = tfd.MultivariateNormalDiag(loc=tf.zeros([2], DTYPE), scale_diag=tf.ones([2], DTYPE))
```

```python
base_dist.parameters
```

```python
bijectors = []
for i in range(num_bijectors):
    bijectors.append(tfb.MaskedAutoregressiveFlow(
        shift_and_log_scale_fn=tfb.masked_autoregressive_default_template(
            hidden_layers=[32,32])))
    bijectors.append(tfb.Permute(permutation=[1, 0]))
# discard last permute layer
flow_bijector = tfb.Chain(list(reversed(bijectors[:-1])))
```

```python
dist = tfd.TransformedDistribution(distribution=base_dist,bijector=flow_bijector)
```

```python
# visualization
x = base_dist.sample(8000)
samples = [x]
names = [base_dist.name]
for bijector in reversed(dist.bijector.bijectors):
    x = bijector.forward(x)
    samples.append(x)
    names.append(bijector.name)
```

```python
session.run(tf.compat.v1.global_variables_initializer())
```

```python
results = session.run(samples)
f, arr = plt.subplots(1, len(results), figsize=(4 * (len(results)), 4))
X0 = results[0]
for i in range(len(results)):
    X1 = results[i]
    idx = np.logical_and(X0[:, 0] < 0, X0[:, 1] < 0)
    arr[i].scatter(X1[idx, 0], X1[idx, 1], s=10, color='red')
    idx = np.logical_and(X0[:, 0] > 0, X0[:, 1] < 0)
    arr[i].scatter(X1[idx, 0], X1[idx, 1], s=10, color='green')
    idx = np.logical_and(X0[:, 0] < 0, X0[:, 1] > 0)
    arr[i].scatter(X1[idx, 0], X1[idx, 1], s=10, color='blue')
    idx = np.logical_and(X0[:, 0] > 0, X0[:, 1] > 0)
    arr[i].scatter(X1[idx, 0], X1[idx, 1], s=10, color='black')
    arr[i].set_xlim([-10, 10])
    arr[i].set_ylim([-10, 10])
    arr[i].set_title(names[i])
```

```python
loss = -tf.reduce_mean(dist.log_prob(x_samples))
train_op = tf.compat.v1.train.AdamOptimizer(1e-4).minimize(loss)
```

```python
session.run(tf.compat.v1.global_variables_initializer())
```

```python
NUM_STEPS = int(2e5)
global_step = []
np_losses = []
for i in range(NUM_STEPS):
    _, np_loss = session.run([train_op, loss])
    if i % 100 == 0:
        global_step.append(i)
        np_losses.append(np_loss)
    if i % int(1e4) == 0:
        print(i, np_loss)
fig = plt.figure(figsize=(10,5))
plt.plot(np_losses)
plt.show()
```

```python
results = session.run(samples)
X0 = results[0]
rows = 2
cols = int(len(results) / 2)
f, arr = plt.subplots(2, cols, figsize=(4 * (cols), 4 * rows))
i = 0
for r in range(rows):
    for c in range(cols):
        X1 = results[i]
        idx = np.logical_and(X0[:, 0] < 0, X0[:, 1] < 0)
        arr[r, c].scatter(X1[idx, 0], X1[idx, 1], s=10, color='red')
        idx = np.logical_and(X0[:, 0] > 0, X0[:, 1] < 0)
        arr[r, c].scatter(X1[idx, 0], X1[idx, 1], s=10, color='green')
        idx = np.logical_and(X0[:, 0] < 0, X0[:, 1] > 0)
        arr[r, c].scatter(X1[idx, 0], X1[idx, 1], s=10, color='blue')
        idx = np.logical_and(X0[:, 0] > 0, X0[:, 1] > 0)
        arr[r, c].scatter(X1[idx, 0], X1[idx, 1], s=10, color='black')
        arr[r, c].set_xlim([-5, 5])
        arr[r, c].set_ylim([-5, 5])
        arr[r, c].set_title(names[i])

        i += 1
```
